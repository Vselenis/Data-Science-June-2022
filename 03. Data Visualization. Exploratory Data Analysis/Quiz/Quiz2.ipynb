```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nuclear_dataset = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv")
nuclear_dataset
```

```python
nuclear_dataset["cost"].mean()
```

```python
north_east_stations = nuclear_dataset["ne"].where(nuclear_dataset["ne"]== 1)
north_east_stations.count()
```

```python
plt.plot(nuclear_dataset.groupby(nuclear_dataset.date).cost.mean())
plt.xlabel("Years")
plt.ylabel("Millions of dollars")
plt.title("Power Plant Cost")
```
