```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage.io
from skimage.io import imread, imshow
```

```python
hamburger = imread("hamburger.jpg")
hamburger.nbytes
```

```python
asd = imshow("hamburger.jpg")
asd
```

```python
red = hamburger[:, :, 0]
green = hamburger[:, :, 1]
blue = hamburger[:, :, 2]
print(red.mean())
print(green.mean())
print(blue.mean())
```

```python
binary_opening = skimage.morphology.binary_opening(blue, selem=skimage.morphology.square(3)).astype(int)
plt.imshow(binary_opening)
plt.show()
```

```python
np.sum(binary_opening)
```
